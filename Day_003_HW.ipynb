1. 專案的目標？ (要解決什麼問題）

為 Apple Music 建立一個名為 ELISE 的全球多語言語義檢索系統，以改善其搜尋系統的Recall並提升搜尋轉化率。最終，這套系統在 A/B 測試中成功降低了 86% 的「無搜尋結果率」，其問題背景為：

- 全球化與語言複雜度：Apple Music 服務全球超過 150 個市場、涵蓋數十種語言，且曲庫每日以數十萬首新歌的速度增長
- 傳統 Token 檢索的局限性：原有的搜尋系統主要使用基於 Token（詞彙符號）的字面匹配索引，這在應對拼寫錯誤（如將 KATSEYE 拼錯成 "gnarly by cat's eye"）、拼音轉換、跨語言查詢以及**功能性/描述性查詢（如搜尋 "party music from the 90s"）**時，容易因字面重疊極低或不存在，導致搜尋失敗並產生「No-Result」的情況
- 長尾查詢的失敗：這些檢索失敗高度集中在Tail queries（低頻但佔了所有獨特查詢的 83%）。因此，新系統旨在不影響熱門查詢的前提下，精準解決這些困難的長尾查詢召回問題

2. 使用的技術是？ (只需知道名稱即可，例如：使用 CNN 卷積神經網路做影像分類)

- Siamese Bi-encoder：微調自 305M 參數的開源模型 GTE-multilingual-base

- Curriculum-scheduled Multi-objective Training：在訓練中，從前期的 InfoNCE 損失函數 逐漸過渡到後期的 Pairwise Hinge Loss with Hardest-in-Batch Negatives

- Matryoshka Representation Learning, MRL

- Approximate Nearest Neighbor, ANN：使用 FAISS 向量索引庫，並透過 OPQ（最佳化乘積量化）與 HNSW 進行加速與壓縮

- Hybrid Retrieval：融合密集語義檢索與傳統的稀疏 Token 索引

- Quantile Distribution Matching：用於融合向量相似度分數與文字匹配分數，使混合分數符合下游排序器預期的分布，避免重新訓練排序模型


3. 資料來源？

- Apple Music Worldwide Search Logs：
  - Query-Item Pairs：從全球搜尋日誌中採樣，當用戶搜尋後產生了成功互動（如播放、加到資料庫、分享等轉化行為）時，即形成 Query 與音樂實體的配對數據
  - Query-Query Pairs：在同一個搜尋工作階段（Session）中，用戶輸入失敗的查詢與其隨後成功修正的查詢配對（如拼寫修正），用來教導模型學習拼寫糾錯

- Apple Music Catalog Entity Metadata：將歌曲、專輯、藝人、歌單、廣播電台等目錄實體的所有可用欄位線性化成結構化文本。包含的資訊有：標題、歌手、曲風、發行資訊（如發行年份或年代描述）、受歡迎程度分桶、歌詞片段（Lyrics Snippet）以及歷史上曾引導至該實體的高頻關聯查詢。 (註：研究團隊曾嘗試加入 LLM 生成的配對、網頁共現字詞以及歌單數據等額外數據源，但發現這些補充數據反而會干擾真實搜尋分布，因此最終生產環境僅使用上述最真實的「搜尋日誌」與「目錄元數據」）
